## 1. Exercises with numbers and simple operations

1. Age in 5 years.
- Create a variable age that is your age is numbers. E.g. 25.
- Create another variable age_in_five_years that is equal to your age times 5.
- Print out the result using print() statement.
2. Age in dog-years.
- Create (or reuse) a variable age that is your age is numbers. E.g. 25.
- Create another variable multiplier that is 7.  
- Create another variable age_in_dog_years that is equal to your age times multiplier.
- Print out the result using print() statement.
3. Average of 5 numbers.
- Create variable average with the average of the numbers.
  - E.g. Take 5 numbers (e.g. 1, 2, 3, 4, 5).
- PS! Parentheses work for precedence also in Python.
- Print out average using the print() statement.


## 2. String manipulation exercises

1. Favorite Hobby.
- Create a variable hobby that stores a favorite hobby as a string.
- Use input() function with prompt: "What is your favorite hobby?"
- Create a variable sentence that combines the phrase "I love " with your hobby.
- Print the sentence using the print() statement.
2. Age and Name Manipulation
- Create a variable name that stores your name as a string.
  - Use input() function with prompt: "What is your name?"
- Create another variable age that stores your age as a string.
  - Use input() function with prompt: "What is your age?"
- Create a variable greeting that combines your name and age into the following message:
  - "Hello, my name is John and I am 55 years old."
- Print out the greeting.


## 3. Boolean and Conditions Exercises

1. Age Check
- Create a variable my_age that stores your age as a number. E.g. 14.
- Create another variable friend_age that stores your friend’s age. E.g. 16.
- Create a variable is_older that checks if your age is greater than your friend’s age.
- Print the result of is_older.
- Create another variable same_age that checks if your age is equal to your friend’s age.
- Print the result of same_age using the print() statement.
2. Passing the Test
- Create a variable my_score that stores your test score as a number. E.g. 85.
- Create another variable passing_score that stores the minimum score needed to pass. E.g. 60.
- Create a variable passed that checks if your score is greater than or equal to the passing score.
- Print the result of passed using the print() statement.
- Create another variable excellent that checks if your score is greater than 90.
- Print the result of excellent.
